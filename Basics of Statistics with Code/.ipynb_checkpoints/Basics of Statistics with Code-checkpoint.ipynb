```python
import math
```

## Mean 


```python
def mean(*args):
    val_sum = sum(args)
    return val_sum / len(args)
```

```python
print(f"Mean : {mean(1,2,3,5,10)}")
```

## Median 



```python
def median(*args):
    if len(args) % 2 == 0:
        i = round((len(args) + 1) / 2)
        j = i-1
        return (args[i] + args[j]) / 2
    else:
        k = round(len(args)/2)
        return args[k]
```

```python
print(f"Median : {median(1,2,12,4,5)}")
```

## Mode



```python
def mode(*args):
    dict_vals = {i: args.count(i) for i in args}
    print(dict_vals)
    max_list = [k for k, v in dict_vals.items() if v == max(dict_vals.values())]
    print(max_list)
    return max_list
```

```python
print(f"Mode : {mode(2,2,2,2,4,4,4,5,6,6,6,6,6,6)}")
```

## Variance


```python
def variance(*args):
    mean_val = mean(*args)
    numerator = 0
    for i in args:
        numerator += (i - mean_val) ** 2
    denominator = len(args) - 1
    return numerator / denominator
```

```python
print(f"Variance : {variance(3,5,6,7,8)}")
```

## Standard Deviation : Standard deviation tells you how spread out the data is.It is a measure of how far each observed value is from the mean.





```python
def standard_deviation(*args):
    return math.sqrt(variance(*args))
```

```python
print(f"Standard Deviation : {standard_deviation(3,3,4,6,7,9)}")
```

## Coefficient Variations


```python
def coefficient_variation(*args):
    return standard_deviation(*args) / mean(*args)
```

```python
print(f"Coefficient Variations {coefficient_variation(2,3,4,5,7,8,9)}")
```

## Covariance : Covariance tells us if two values moving in the same direction.
1. /> 0 : Moving Together
2. < 0 : Moving Opposite
3. = 0 : Independent


```python
def covariance(*args):
    list_1 = [i[0] for i in args]
    print(list_1)
    list_2 = [i[1] for i in args]
    print(list_2)
    list_1_mean = mean(*list_1[0])
    list_2_mean = mean(*list_2[0])
    numerator = 0
    if len(list_1[0]) == len(list_2[0]):
        for i in range(len(list_1[0])):
            numerator += (list_1[0][i] - list_1_mean) * (list_2[0][i] - list_2_mean)
            denominator = len(list_1[0]) - 1
            return numerator / denominator
        else:
            print('Error')

    
```

```python
salary_exper_list = [[1333,4564,7554,76554,8456,9543],[1,2,3,4,5,6]]
print(f"Covariance : {covariance(salary_exper_list)}")
```
