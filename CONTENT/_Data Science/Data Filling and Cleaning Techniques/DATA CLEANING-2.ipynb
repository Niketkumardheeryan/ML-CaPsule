```python
# percentage data missing
null_var=df.isnull().sum()/df.shape[0]*100
null_var
```

```python
# list columns having more than 20% missing data
drop_columns = null_var[null_var > 20].keys()
drop_columns
```

```python
# drop above columns
df2_drop_col=df.drop(columns=drop_columns)
df2_drop_col.shape
```

```python
# select rows having numeric dataset only
df_num=df2_drop_col.select_dtypes(include=['int64','float64'])
df_num.head()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df_num.isnull())
```

```python
# show rows and columns having any NaN or missing value
df_num[df_num.isnull().any(axis=1)]
```

```python
df_num.isnull().sum()
```

```python
# names of columns having missing values
miss_num_var=[var for var in df_num.columns if df_num[var].isnull().sum()>0]
miss_num_var
```

```python
# data distribution to predict what to use median or mode or mean

plt.figure(figsize=(10.,10))
sns.set()
for i,var in enumerate(miss_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var], bins=20, kde_kws={ 'linewidth' : 5, 'color': 'r'})
```

```python
# normal distribution -> can be mirror halved  (mean/mode/median) fig.1
# skewed distribution -> fig. 2,3  (median near to top / mean mid of other half side)
```

```python
# fill the missing values
# fill with the MEAN values
df4_num_mean=df_num.fillna(df_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
# check data distribution of original data and cleaned data
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(miss_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var], bins=20, kde_kws={ 'linewidth' : 5, 'color': 'r'},label="original")
    sns.distplot(df4_num_mean[var], bins=20, kde_kws={ 'linewidth' : 5, 'color': 'g'},label="cleaned")
    plt.legend()
```

```python
# fill the missing values
# fill with the MEDIAN values
df5_num_median=df_num.fillna(df_num.median())
df5_num_median.isnull().sum().sum()
```

```python
# check data distribution of original data and cleaned data
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(miss_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df_num[var], bins=20, kde_kws={ 'linewidth' : 8, 'color': 'r'},label="original")
    sns.distplot(df4_num_mean[var], bins=20, kde_kws={ 'linewidth' : 5, 'color': 'g'},label="cleaned mean")
    sns.distplot(df5_num_median[var], bins=20, kde_kws={ 'linewidth' : 2, 'color': 'k'},label="cleaned median")
    plt.legend()
```

```python
# box plot 
# to check  the outlers
for i, var in enumerate(miss_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var]) 
    plt.subplot(3,1,3)
    sns.boxplot(df5_num_median[var])
```

```python
# concat the data 
df_concat=pd.concat([df_num[miss_num_var],df4_num_mean[miss_num_var],df5_num_median[miss_num_var]],axis=1)
```
