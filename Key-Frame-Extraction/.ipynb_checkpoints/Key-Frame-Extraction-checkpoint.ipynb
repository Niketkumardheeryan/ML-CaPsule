```python
import cv2#importing cv2 library
import numpy as np#importing numpy library
import os
```

```python
def framing(video):#defining a small function named"framing" with a parameter "i" that's supposed to be provided for reading the video
    fr = []#creating an empty list named fr
    fr_pre=[]#creating an empty list named fr_pre
    cap = cv2.VideoCapture(video)#reading the video file
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    while (cap.isOpened()):
        ret,frame = cap.read()
        if ret == True:#checking for presence of frames
            cv2.imshow("original",frame)#displaying the frames
            grayed = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
            canned = cv2.Canny(grayed,320,320)
            fr.append(frame)
            fr_pre.append(canned)
            cv2.imshow("Grayed",grayed)
            cv2.imshow("Canned",canned)
            k = cv2.waitKey(10) & 0XFF
            if k == ord('q'):
                break
        else:
            break
    cap.release()
    
    cv2.destroyAllWindows()      
    return fr_pre,fr,fps     
```

```python
frames,ogframes,fps = framing("sample1.mp4")
cv2.destroyAllWindows()
fps
```

```python
diff = []
for i in range(0,len(frames)-1):
    print(frames[i],frames[i+1])
    diff.append(cv2.absdiff(frames[i],frames[i+1]))
```

```python
#Calculating the Global Threshold Value
mn = np.mean(diff)
st_d = np.std(diff)
print(mn,st_d)
a = 4 
ts = mn + (a * st_d)
print('The threshold==>',ts)
```

```python
#Local Threshold Value
a_fr = []
for i in range(len(diff)):
    mn = np.mean(diff[i])
    st_d = np.std(diff[i])
    fr_ts = mn + (4*st_d)
    print(i,fr_ts)
    a_fr.append([i,fr_ts])
```

```python
#Extracting the Timestamps and KeyFrames Extraction
imp_fr = []
timestamps = []
for i,ac_tr in(a_fr):
    if ac_tr >= ts:
        print(i,ac_tr)
        timestamps.append([i,i/fps])
        imp_fr.append([i,ac_tr])
```

```python
#Final KeyFrames and TimeStamp Appending
key_fr = []
print(len(ogframes))
print(len(timestamps))
cnt=0
for i,_ in imp_fr:
    key_fr.append([ogframes[i],timestamps[cnt][1]]) 
    cnt+=1
    print(ogframes[i],i)
print(key_fr)
```

```python
#This code helps to check how many frames are being extracted
for i in range(len(key_fr)):
    cv2.imshow('keys',key_fr[i][0])
    print("Frame no==>",i,"\t Time Stamps==>",key_fr[i][1])
    k = cv2.waitKey(100) & 0xFF
    if k == ord('q'):
        break
cv2.destroyAllWindows()
```

```python
#Importing the YoloV3 module
net=cv2.dnn.readNet(r"C:\Users\malpu\OneDrive\Desktop\Key-Frame-Extraction\yolov3.weights",r"C:\Users\malpu\OneDrive\Desktop\Key-Frame-Extraction\yolov3.cfg")
classes=[]
with open (r"C:\Users\malpu\OneDrive\Desktop\Key-Frame-Extraction\coco.names.txt") as f :
    classes = [line.rstrip() for line in f.readlines()]
layer_names = net.getLayerNames()
outputlayers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]
```

```python
#Function that returns the Object Count in the Frame Images 
def getObjCnt(mypath,image):
    #loading image
    path=os.path.join(mypath, image)
    img = cv2.imread(path)
    img = cv2.resize(img,None,fx=1,fy=1)
    height,width,channels = img.shape
    blob = cv2.dnn.blobFromImage(img,1/255.0,(416,416),(0,0,0),True,crop=False)
    net.setInput(blob)
    outs = net.forward(outputlayers)
    class_ids=[]
    confidences=[]
    boxes=[]
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]

            if confidence > 0.7:
                #onject detected
                center_x= int(detection[0]*width)
                center_y= int(detection[1]*height)
                w = int(detection[2]*width)
                h = int(detection[3]*height)
                x=int(center_x - w/2)
                y=int(center_y - h/2)
                boxes.append([x,y,w,h])
                confidences.append(float(confidence))
                class_ids.append(class_id)
                
    indexes = cv2.dnn.NMSBoxes(boxes,confidences,0.4,0.6)

    colors= np.random.uniform(0,255,size=(len(boxes),3))
    font = cv2.FONT_HERSHEY_PLAIN
    for i in range(len(boxes)):
        if i in indexes:
            x,y,w,h = boxes[i]
            label = str(classes[class_ids[i]])
            color = colors[i]
            cv2.rectangle(img,(x,y),(x+w,y+h),color,2)
            cv2.putText(img,label,(x,y+30),font,1,(255,255,255),2)
    cv2.imshow("Image",img)
    k = cv2.waitKey(1) & 0xFF
    if k == ord('q'):
        return -1
    return len(indexes)
```

```python
#Here we are just passing all the KeyFrames from the saved Directory to the getObjCnt Function
all_files=[]
ObjCnts={}
path="C:\\Users\\malpu\\OneDrive\\Desktop\\Key-Frame-Extraction\\Key_Frames\\"
for file in os.listdir(path):
    if file.endswith(".jpg"):
        all_files.append(file)
for image in all_files:
    x=getObjCnt(path,image)
    if x ==-1:
        cv2.destroyAllWindows()
        break
    print(image,"==>",x)
    ObjCnts[image]=x
```
