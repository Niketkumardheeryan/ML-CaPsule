Building the logistic regression model

```python
def initialise(dim):
        '''Initialise the parameter values and bias values'''
        w = np.zeros((dim,1))
        b = 0
        print("w",w.shape)
        return w,b
        

def sigmoid(z):
    '''Compute sigmoid function'''
    sig = 1/(1+np.exp(-z))
    
    return sig

def cost_fn(w,b,x_train,y_train):
    '''Written cost function for both parameters and bias'''
    m = x_train.shape[1]
   
    
    A = sigmoid(np.dot(w.T,x_train)+ b)
   
    cst_fn = (-1)*(np.sum((np.log(A.T))*y_train+ (np.log(1-A.T))*(1-y_train)))/m
    
    
    

    dw = np.dot(x_train,(A-y_train).T)/m
    db = np.sum(A-y_train)/m
  

    gradient = {"dw":dw, "db": db}
   
    return cst_fn, gradient

    


```

Implementing Gradient descent

```python
def gradient_descent(w,b,x_train,y_train,epochs,learning_rate):
    
    for i in range(epochs):
        costs = []
        cost,gradient = cost_fn(w,b,x_train,y_train)

        dw = gradient["dw"]
        db= gradient["db"]

        w = w- (learning_rate*dw)
        b= b - (learning_rate*db)

        if i%50==0:
            costs.append(cost)
    parameters = {"w":w,"b":b}
    gradient = {"dw":dw,"db":db}
    

    return  parameters, gradient, costs
    


```

Writing the predicting function 

```python
def predict(w,b,x_test):
    m = x_test.shape[1]
    y_prediction = np.zeros((1,m))
    w = w.reshape(x_test.shape[0],1)
    
    A = sigmoid(np.dot(w.T,x_test)+b)

    for i in range(A.shape[1]):
        if A[0,i]>=0.5:
            y_prediction[0,i]=1
        else:
            y_prediction[0,i]=0
    
    return y_prediction
```

Combining preceeding functions to make the logistic regression model 

```python
def logistic_regression(x_train, y_train,x_test,y_test,epochs=100,learning_rate=3):
    w,b = initialise(x_train.shape[0])

    parameters,gradient,costs= gradient_descent(w,b,x_train,y_train,epochs,learning_rate)
    w = parameters["w"]
    b = parameters["b"]
    
    y_pred_train = predict(w,b,x_train)
    y_pred_test = predict(w,b,x_test)

    accuracy_train= (100 - np.mean(np.abs(y_pred_train-y_train)*100))
    accuracy_test =(100 - np.mean(np.abs(y_pred_test-y_test)*100))

    print("Accuracy on training data", accuracy_train)
    print("Accuracy on test data", accuracy_test)
    return parameters
```
