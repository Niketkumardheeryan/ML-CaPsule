# What is Interquartile Range?

IQR is used to measure variability by dividing a data set into quartiles. The data is sorted in ascending order and split into 4 equal parts. Q1, Q2, Q3 called first, second and third quartiles are the values which separate the 4 equal parts.


• Q1 represents the 25th percentile of the data.

• Q2 represents the 50th percentile of the data.

• Q3 represents the 75th percentile of the data.




IQR is the range between the first and the third quartiles namely Q1 and Q3:

**IQR = Q3 – Q1**

The data points which fall below **Q1 – 1.5IQR** or **above Q3 + 1.5IQR** are outliers..

## Visualizing the dataset for proper understanding

```python
plt.scatter(df['Height'], df['Weight'], s=5, alpha=0.5, marker='.')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs. Weight with Customized Markers')
plt.show()
```

## For detecting outliers in Weight column of the dataset.

```python
sns.boxplot(data=df.Weight)
plt.show()
```

```python
plt.hist(df.Weight, bins=20, rwidth=0.8)
plt.xlabel('Weight')
plt.ylabel('Counts')
plt.show
```

```python
sns.kdeplot(df['Weight'])
plt.title('Density Plot of Weight')
plt.xlabel('Weight')
plt.show()
```

```python
Q1_weight = df.Weight.quantile(0.25)
Q3_weight = df.Weight.quantile(0.75)
Q1_weight,Q3_weight
```

```python
IQR = Q3_weight - Q1_weight
IQR
```

```python
lower_limit_weight = Q1_weight - (1.5*IQR)
upper_limit_weight = Q3_weight + (1.5*IQR)
lower_limit_weight,upper_limit_weight
```

```python
df1 = df[(df.Weight>upper_limit_weight) | (df.Weight<lower_limit_weight)]
df1
```
