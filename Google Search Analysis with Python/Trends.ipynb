```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
pytrends = TrendReq()
```

```python
pytrends.build_payload(kw_list=["Electric Vehicle"])
data = pytrends.interest_by_region()
data = data.sort_values(by="Electric Vehicle", ascending=False)
data = data.head(20)
print(data)
```

```python
#Plotting a Chart to visualize the trends countrywise.
data.reset_index().plot(x="geoName", y="Electric Vehicle", 
                        figsize=(20,5), kind="bar")
plt.show()
```

```python
#Plotting a graph to visualize the searches year over year.
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Electric Vehicle'])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(15,10))
data['Electric Vehicle'].plot()
plt.style.use("ggplot")
plt.title('Google Searches for Blockchain', fontweight='light')
plt.xlabel('Year')
plt.ylabel('Trend Score')
plt.show()
```

```python
#Conclusion-
#As we can see that year on year searches related to electric vehicle is continuously 
#increaing and searches of electric vehicle is predominantly high in developing countries in comparision to develop countries.
```
