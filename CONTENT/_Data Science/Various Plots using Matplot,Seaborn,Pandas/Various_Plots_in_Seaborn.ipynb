# KDE Plot

```python
#Using normal matplotlib hist() function we will get a histogram.
#Rather than a histogram, we can get a smooth estimate of the distribution 
#using a kernel density estimation, which Seaborn does with sns.kdeplot
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

for col in 'xy':
    sns.kdeplot(data[col], shade=True)
```

# Distplot

```python
#We can easily combine histogram and KDE using distplot() function
sns.distplot(data['x'])
sns.distplot(data['y'])
```

# **Categorical** **Plots**

# Bar Plot

```python
#It is an example of bivariate analysis.
#On the x-axis, we have a categorical variable and on the y-axis, we have a continuous variable.
sns.barplot(x=df['Sex'],y=df['Fare'])

#From this we interpret that the average fare value of female passengers is higher than that of male
#passengers
```

# Count Plot

```python
# It is is used to count the occurence of a categorical variable
#It is also an example of univariate analysis


sns.countplot(df['Survived'])
```

# Box Plot

```python
#It is a 5 point summary plot. It gives the information 
#about the maximum, minimum, mean, first quartile, and third quartile of a continuous variable. 
#Also, it equips us with knowledge of outliers.

sns.boxplot(x=df['Age'],y=df['Sex'])
```

# Violin Plot

```python
#Similar to box plot but along with that it provides some supplementary information about variables

sns.violinplot(x=df['Age'],y=df['Sex'])
```

# **Advanced Plots**



*   Strip Plot
*  Swarm Plot






#Strip Plot

```python
#It is used to plot a graph between a continuous variable and categorical variable
#Here fare is a continuous variable and Survived is a Categorical variable as it can be only 0 or 1

sns.stripplot(y=df['Fare'],x=df['Survived'])
```

# Swarm Plot

```python
#It is a combination strip plot and violin plot
# It provides respective distribution as an additional feature

sns.swarmplot(y=df['Age'],x=df['Survived'])
```

# **Matrix Plots**

*   Heat Map
*   Cluster Map



# Heat Map

```python
# These are used to understand relations between various attributes.
#In cas of large datasets it is difficult to traverse through all values and find relation
# So heatmap eases the visualization
# The boxes with value 1 are the boxes which means they are exactly same and hence the value is 1
# Similarly it goes on

sns.heatmap(df.corr(),annot=True,cmap='viridis')
```

```python
# Heatmaps are also used to find the missing patterns
# Here the yellow dash represent NULL value or the value is missing from those positions in that column

sns.heatmap(df.isnull(),yticklabels=False,cbar='False',cmap='viridis')
```

# Cluster Map

```python
#The x-label and y-label are the same but they harmonized differently. 
#That is because they are grouped according to their similarity.
#The flow-chart like structure at the top and left describe their degree of similarity.
#Cluster maps use Hierarchical clustering to form different clusters.

sns.clustermap(df.corr(),annot=True,cmap='viridis')
```

# Facet Grid

```python
# In case we want a single visualization of different categories based on one continuous variable
# facet Grid is used
#Here we wanted the Age distribution of males and females in various classes 

sns.FacetGrid(df,col='Pclass',row='Sex').map(sns.distplot,'Age')
```

# Regression Plot

```python
# This is an advanced plot which scatter plots categorical variables along with a linear fitting line.

sns.lmplot(x='Age',y='PassengerId',data=df,hue='Pclass')
```
