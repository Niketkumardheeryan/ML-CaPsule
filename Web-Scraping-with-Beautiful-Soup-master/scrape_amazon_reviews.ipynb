```python
""" This code scrapes customer rewiews from an amazon product and organizes them in a .csv file
"""

import requests
from bs4 import BeautifulSoup
from datetime import datetime
import logging # For logging the progress of the script

# Logging configuration
logging.basicConfig(level=logging.INFO)

# Headers for the HTTP request to the website
headers = {'User-Agent' : 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36'}

# URL of the page to scrape
URL = "https://www.amazon.in/BULLMER-Mens-Slim-Shirt-BUL-BSP015_013_017-L_Multicolored_Large/product-reviews/B0892SZX7F/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews"

logging.info(f"Fetching URL: {URL}") # Log the URL being fetched
```

```python
# Send a GET request to the URL
resp = requests.get(URL, headers=headers)
print(resp.status_code) #check if the request was successful, 200 means success
```

```python
page_html = resp.text # Get the HTML content of the page
print(page_html)
```

```python
# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(page_html, "lxml")

print(soup.prettify())
```

```python
# extract all reviews in class: a-section celwidget
reviews = soup.find_all("div", {"class": "a-section celwidget"})
print(reviews)
```

```python
# List to hold the extracted review data
all_results = []

# Iterate through each review element and extract the required information
for review in reviews:
    review_text = review.find("span", {"class": "a-size-base review-text review-text-content"}).get_text().strip() # Extract the review text
    review_date = review.find("span", {"class": "review-date"}).get_text() # Extract the review date
    stars = review.find("span", {"class": "a-icon-alt"}).get_text().strip() # Extract the number of stars
    
    # Append the extracted data to the results list
    all_results.append({
        "review_text": review_text,
        "review_date": review_date,
        "review_stars": stars,
    })

for result in all_results:
    print(result)
```
