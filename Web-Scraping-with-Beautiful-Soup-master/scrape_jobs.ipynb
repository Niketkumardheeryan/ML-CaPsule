```python
"""this file is a demo on web-scrapping on the website https://realpython.github.io/fake-jobs/  which is an example site with fake job postings that you can freely scrape to train your skills
"""

import requests 
from bs4 import BeautifulSoup

url = "https://realpython.github.io/fake-jobs/" 
page = requests.get(url) #get the page using requests
print(page.status_code) #returns 200 if successful
```

```python
soup = BeautifulSoup(page.content, 'html.parser') #parse the html content of the page

print(soup.prettify()) #prints the html content of the page
```

```python
#narrow down the search to the div element with the id ResultsContainer
results = soup.find(id="ResultsContainer")
print(results.prettify())
```

```python
#find all the job elements in class: card-content 
job_elements = results.find_all("div", class_="card-content")
for job_element in job_elements:
    print(job_element.prettify(), end="\n"*2) #print the job elements with two new lines in between
```

```python
#narrowing down the data required from the page by using the find() method and unique class names
for job_element in job_elements:
    title_element = job_element.find("h2", class_="title")
    company_element = job_element.find("h3", class_="company")
    location_element = job_element.find("p", class_="location")
    print(title_element)
    print(company_element)
    print(location_element)
    print()
```

```python
#extracting the relevant text from the elements
for job_element in job_elements:
    title_element = job_element.find("h2", class_="title") #extracting the title of the job
    company_element = job_element.find("h3", class_="company") #extracting the company name
    location_element = job_element.find("p", class_="location") #extracting the location of the job

    print(title_element.text.strip()) #strip() removes any leading or trailing whitespace
    print(company_element.text.strip())
    print(location_element.text.strip())
    print()
    #extracting the link to apply for the job
    links = job_element.find_all("a")
    link = links[1]["href"] #the second link is the link to apply for the job
    print(f"Apply here: {link}\n")
    
```

```python
#organizing the output in a csv file
import csv

with open("jobs_available.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile) #create a csv writer object

    writer.writerow(["Title", "Company", "Location", "Link to Apply"]) #write the header of the csv file
    
    #loop through the job elements and write the requiered data to the csv file
    for job_element in job_elements:
        title_element = job_element.find("h2", class_="title") #extracting the title of the job
        company_element = job_element.find("h3", class_="company") #extracting the company name
        location_element = job_element.find("p", class_="location") #extracting the location of the job
        apply_link = job_element.find_all("a")[1]["href"] #extracting the link to apply for the job

        writer.writerow([title_element.text.strip(), company_element.text.strip(), location_element.text.strip(), apply_link]) #write the data to the csv file
print('Done')
```
