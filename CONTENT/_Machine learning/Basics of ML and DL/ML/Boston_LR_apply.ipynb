```python
# train-test split

from sklearn import model_selection

x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y,test_size=0.2)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# apply  model

from sklearn.linear_model import LinearRegression

alg1 = LinearRegression() # creates an object of LinearRegression
```

```python
# training of model

alg1.fit(x_train, y_train)
```

```python
# predicting the output for test data

y_predict = alg1.predict(x_test)
```

```python
# compare y_predict and y_test

import matplotlib.pyplot as plt

plt.scatter(y_test, y_predict)
plt.axis((0,50,0,50))
plt.show()
```

```python
print("Train Accuracy: ",alg1.score(x_train, y_train))
print("Test Accuracy: ",alg1.score(x_test, y_test))
```
