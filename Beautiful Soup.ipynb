```python
from bs4 import BeautifulSoup
import requests as r
```

```python
source= r.get('https://coreyms.com').text     # getting the html file of a web page

```

```python
soup=BeautifulSoup(source,'lxml')           # parse as a lxml

soup                                         
```

```python
soup.title
```

```python
soup.prettify()       # well oriented 
```

```python
print(soup.prettify())             #looks pretty to use prettify()
```

```python
print(soup.script.prettify())
```

```python
# finfding a particular tags

soup.title                                  #with tag          , will only get the first one title we create 
soup.title.text                             #without tag



```

```python
link=soup.find('link')
```

```python
# splitting the link from '/ 'symbol 
link.split('/')[4]
```

```python
# search a particular tag we needed

style=soup.find('style')
print(style)
```

## Nested tags find 

```python
## for nested tags use this method

nested=soup.find('div',class_='site-container')

print(nested.find('g',class_='site-avatar-foreground').prettify())
```

```python
soup.find('style')                           # find() return a list 
```

## To find all tags with same names

```python
# for Looping over a articles or tag

for i in soup.find_all('link'):
    print(i)
    
                                                                       # all link  tags  will be printed
    
```

## save the data or text into Different types of files 

```python
#into csv files
import csv

csv_file=open('scrapping_file.csv','w')
```

```python
csv_writer = csv.writer(csv_file)

csv_writer.writerow(['links'])              # Means simply write the col names of csv files
```

```python
links=soup.find_all('link')
print(links)
csv_writer.writerow(links)
    
    
            
```

```python
csv_file.close()
```
