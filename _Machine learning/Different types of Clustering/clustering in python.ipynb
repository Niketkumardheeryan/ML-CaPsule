**DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**

**Hierarchical Clustering**

```python
from sklearn.cluster import AgglomerativeClustering
agg_clustering = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
labels2 = agg_clustering.fit_predict(df_4)
```

```python
df_4['clus'] = labels2
```

```python
sb.scatterplot(df_4['PetalLengthCm'],df_4['PetalWidthCm'],hue = df_4['clus'])
plt.show()
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20, 7))  
plt.title("Dendrograms")  
sc.dendrogram(sc.linkage(df_4, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Euclidean distance')
```

**Partitioning Clustering**

**K-means clustering**

```python
cluster_range = range(2,10)
cluster_wss = []
from sklearn.cluster import KMeans
for num_cluster in cluster_range:
    cluster = KMeans(num_cluster,random_state=25)
    cluster.fit(x)
    cluster_wss.append(cluster.inertia_)
```

```python
cluster_df = pd.DataFrame({'num_cluster':cluster_range,'cluster_wss':cluster_wss})
cluster_df
```

```python
plt.xlabel('no. of cluster')
plt.ylabel('WSS')
plt.plot(cluster_df['num_cluster'],cluster_df['cluster_wss'],marker = '*')
```

```python
Kclusters = KMeans (n_clusters = 3,random_state = 25)
Kclusters.fit(df_5)
df_5['clus'] = Kclusters.predict(df_5)
```

```python
sb.scatterplot(df_5['PetalLengthCm'],df_5['PetalWidthCm'],hue = df_5['clus'])
plt.show()
```
