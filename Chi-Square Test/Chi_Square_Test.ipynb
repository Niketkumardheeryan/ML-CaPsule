```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the CSV file
df = pd.read_csv('/content/drive/MyDrive/GirlsScriptOpenSource/ml|RepoProjects/Chi-Square Test /dataset.csv')
```

```python
# Plot the distribution of Gender
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Gender')
plt.title('Distribution of Gender')
```

```python
# Plot the distribution of Preference
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Preference')
plt.title('Distribution of Preference')
plt.tight_layout()
plt.show()
```

```python
# Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Preference'])
```

```python
# Plot the heatmap of the contingency table
plt.figure(figsize=(8, 6))
sns.heatmap(contingency_table, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Contingency Table Heatmap')
plt.show()
```

```python
# Step 3: Apply the Chi-Square Test for Independence

# Perform the Chi-Square Test for Independence
chi2, p, dof, expected = chi2_contingency(contingency_table)
```

```python
print(f"Chi-Square Statistic: {chi2}")
print(f"P-Value: {p}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

```python
# Interpret the p-value
alpha = 0.05
if p < alpha:
    print("There is a significant association between Gender and Preference.")
else:
    print("There is no significant association between Gender and Preference.")
```
