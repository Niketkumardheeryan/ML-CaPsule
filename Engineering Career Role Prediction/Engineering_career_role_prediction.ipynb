```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report, confusion_matrix

label_encoder = LabelEncoder()
df['suggested_job_roll'] = label_encoder.fit_transform(df['suggested_job_roll'])
```

```python
label_mapping = {index: label for index, label in enumerate(label_encoder.classes_)}

# Print the label mappings
print("Label Encoding Mapping:")
for encoded_value, label in label_mapping.items():
    print(f"{encoded_value} - {label}")
```

# **Random Forest**

```python
# Features and target variable
X = df.drop(columns=['suggested_job_roll'])
y = df['suggested_job_roll']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Initialize and train the model
model_rf = RandomForestClassifier(random_state=42)
model_rf.fit(X_train, y_train)
# Make predictions
y_pred = model_rf.predict(X_test)

# Calculate metrics
accuracy_rf = accuracy_score(y_test, y_pred)
print("accuracy = ",accuracy_rf)
precision_rf = precision_score(y_test, y_pred, average='weighted')
print("Precision = ",precision_rf)
recall_rf = recall_score(y_test, y_pred, average='weighted')
print("Recall = ",recall_rf)
f1score_rf = f1_score(y_test, y_pred, average='weighted')
print("F1 Score = ",f1score_rf)

# Print classification report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot Confusion Matrix
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
import joblib
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load the Logistic Regression model
model = joblib.load('logistic.pkl')

# Define the job role mapping
job_role_mapping = {
    0: 'Artificial Intelligence Engineer',
    1: 'Cloud Solutions Architect',
    2: 'Cybersecurity Analyst',
    3: 'Data Scientist',
    4: 'Database Administrator',
    5: 'IT Manager',
    6: 'Network Administrator',
    7: 'Software Developer',
    8: 'Systems Analyst',
    9: 'Web Developer'
}

def preprocess_input(data):
    df = pd.DataFrame([data])
    categorical_columns = ['can work long time before system?', 'self-learning capability?', 'talent tests taken?', 'higher education?']
    label_encoders = {}

    for col in categorical_columns:
        if col in df.columns:
            le = LabelEncoder()
            df[col] = le.fit_transform(df[col])
            label_encoders[col] = le

    return df

# Predict function
def predict_job_role(user_input):
    preprocessed_input = preprocess_input(user_input)
    prediction_numeric = model.predict(preprocessed_input)[0]
    predicted_job_role = job_role_mapping[prediction_numeric]
    return predicted_job_role

# Function to take user input
def get_user_input():
    data = {}
    data['Percentage in Operating Systems'] = int(input('How many marks did you get in Operating Systems (OS) (0-100) ? '))
    data['Percentage in Design and Analysis of Algorithms'] = int(input('How many marks did you get in Design and Analysis of Algorithms (DAOA) (0-100) ? '))
    data['Percentage in Principle of Programming Languages'] = int(input('How many marks did you get in Principle of Programming Languages (POPL) (0-100) ? '))
    data['Percentage in Software Engineering'] = int(input('How many marks did you get in Software Design and Modelling (SDAM) (0-100) ? '))
    data['Percentage in Computer Networks'] = int(input('How many marks did you get in Computer Networks (CN) (0-100) ?'))
    data['Percentage in Electronics Subjects'] = int(input('How many marks did you get in Electronic Subjects (0-100) ?'))
    data['Percentage in Computer Architecture'] = int(input('How many marks did you get in Computer Organization and Architecture (COA) (0-100) ? '))
    data['Percentage in Mathematics'] = int(input('How many marks did you get in Mathematics (0-100) ?'))
    data['Percentage in Communication Skills'] = int(input('How many marks did you get in Communication Skills (0-100) ?'))
    data['Hours working per day'] = int(input('Enter the working Hours per day you want : '))
    data['hackathons'] = int(input('Enter the number of hackathons is which you have participated: '))
    data['coding skills rating'] = int(input('Enter your coding skills rating (0-100): '))
    data['public speaking points'] = int(input('Enter your Public speaking rating (0-100): '))
    data['can work long time before system?'] = input('can you work for long time in front of Computer System ? (Yes/No): ')
    data['self-learning capability?'] = input('Are you ready to self-learn the new Tech stacks and technologies ? (Yes/No): ')
    data['Extra-courses did'] = int(input('Enter the number of Extra courses you did: '))
    data['certifications'] = int(input('Enter the total number of certifications: '))
    data['workshops'] = int(input('Enter the total number of workshops: '))
    data['talent tests taken?'] = input('Did you take the talent test ? (Yes/No): ')
    data['percentage in olympiads'] = int(input('Enter the percentage in olympiads you got : '))
    data['reading and writing skills [1-5]'] = int(input('Enter your reading and writing skills [1-5]: '))
    data['memory capability score [1-10]'] = int(input('Enter your memory capability score [1-10]: '))
    data['higher education?'] = input('Do you want to persue Higher education? (Yes/No): ')
    return data

# Main function to get user input and predict job role
def main():
    user_input = get_user_input()
    predicted_job_role = predict_job_role(user_input)
    print(f"Predicted Job Role: {predicted_job_role}")

if __name__ == "__main__":
    main()
```
