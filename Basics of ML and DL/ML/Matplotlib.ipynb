```python
# plot function of x**3

import numpy as np
x = np.array([1,2,3,4])
y = x**3
plt.plot(x,y,"g--")
plt.show()
```

```python
# the above graph is not looking like x**3 so
x = np.arange(1,5,0.1)  # here we have used arange() in which 0.1 is the step factor of increment
y = x**3
plt.plot(x,y,"y-.")
plt.show()
```

```python
# customizing graphs

x = np.arange(1,5,0.1) 
y = x**3
plt.plot(x,y, color="brown", marker="o") # change the colour of your plots by adding a color parameter in the plot function
plt.show()
```

```python
plt.plot(x,y, color="silver",linewidth=5) #the linewidth parameter adjusts the width of our plot line
plt.show()
```

```python
x = np.arange(1,5,0.1)
y1 = x**2
y2 = x**3

plt.plot(x,y1, color="green", label="x^2",linewidth=5)
plt.plot(x,y2, color="brown",label="x^3",marker="o") #marker parameter will shape your plot points like circular, triangular

plt.title("Graphs",fontsize=20) # title() function will be used to provide the title for the graph
plt.xlabel("x",fontsize=18) # To give a label to the x-axis, use the xlabel function and give it the label name
plt.ylabel("y",fontsize=18) # fontsize parameter to it to adjust the size of the text.
plt.grid() # visualize your graph through a grid by using the grid() function.
plt.legend() # make our graph more descriptive by using a legend (scale) to the graph
plt.axis([0,5,0,100]) # set the axis length for both your axes by specifying the axis() function and giving it the start and end points for both axes

plt.text(4,60,"Text", fontsize=18) 
# to add some text to your graph at some random place in it The text() function is used You need to specify the coordinates where you want to add your text and the content you want to be displayed.

plt.show()
```

```python
# pie chart

# Pie charts can be plotted using the pie() function. We need to specify the sizes of each portion in the chart in a list.
# to make the chart round use the axis() function and passing it the parameter equal.
# We can pass it the colors parameter and specify the colours of each section.
# we can also give each section labels using the labels parameter.
# to display information on the chart itself, We can display the percentage of each section of the pie by giving the 
# parameter autopct. The autopct parameter needs a function (in the form of a string) that demonstrates 
# to display the percentage on the chartfor ex, %.2f%% will display the percentages with two digits to the right of decimal. 
# And it is not necessary that just the percentages can be displayed on the pie chart; 
# we can pass any function we wish into the autopct parameter and itll decide the values on its own.
# You must have seen in several places that sometimes, for more interesting insight, 
# a certain part of the pie chart is usually shown to be slightly out of the pie  a way of highlighting it. 
# by passing the explode parameter and giving it the values by which each section should be pushed out of the pie chart. 
# A value of 1 indicated that the section is to be pushed entirely out of the pie.
# the counterclock parameter can be used to specify fractions direction, clockwise or counter-clockwise.
# The wedgeprops parameter takes in a dictionary of arguments passed to the wedge objects making the pie. 
# For example, you can pass in wedgeprops = {'linewidth': 3} to set the width of the wedge border lines equal to 3.
```

```python
# simplest pie chart
sizes = [3,4,6,2]
plt.pie(sizes)
plt.show()
```

```python
# to make chart round use axis('equal')

sizes = [3,4,6,2]
plt.pie(sizes)
plt.axis('equal')
plt.show()
```

```python
# to change colors use colors in plt()

sizes = [3,4,6,2]
colors = ['blue','purple','pink','orange']
plt.pie(sizes, colors=colors)
plt.show()
```

```python
# to add labels and title

sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','purple','pink','orange']
plt.pie(sizes, colors=colors, labels=labels)
plt.title("split among classes")
plt.show()
```

```python
# to display information on the chart itself

sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','purple','pink','orange']
plt.pie(sizes, colors=colors, labels=labels, autopct='%.2f%%') 
# %.2f means point ke baad 2 digits chahiye and %% means that we want % sign  
plt.title("split among classes")
plt.show()
```

```python
# to highlight a particular class use explode

explode = [0.1,0,0.2,0]
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','purple','pink','orange']
plt.pie(sizes, colors=colors, labels=labels, autopct='%.2f%%', explode=explode) 
# %.2f means point ke baad 2 digits chahiye and %% means that we want % sign
plt.title("split among classes")
plt.show()
```

```python
# by default the order of everything is anticlockwise, to change the order use parameter counterclock in pie()

explode = [0.1,0,0.2,0]
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','purple','pink','orange']
plt.pie(sizes, colors=colors, labels=labels, autopct='%.2f%%', explode=explode, counterclock=False) 
plt.title("split among classes")
plt.show()
```

```python
# by default the start angle is 0, to change the start angle use startangle parameter in pie() 

explode = [0.1,0,0.2,0]
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','purple','pink','orange']
plt.pie(sizes, colors=colors, labels=labels, autopct='%.2f%%', explode=explode, startangle=100) 
plt.title("split among classes")
plt.show()
```

```python
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','purple','pink','orange']
plt.pie(sizes, colors=colors, labels=labels, autopct='%.2f%%',startangle=100, counterclock=False)
plt.title("split among classes")
plt.show()
```

```python
# to highlight a particular class use explode

explode = [0.1,0,0.2,0]
sizes = [3,4,6,2]
labels = ["A","B","C","D"]
colors = ['blue','purple','pink','orange']
plt.pie(sizes, colors=colors, labels=labels, autopct='%.2f%%', explode=explode, shadow=True, wedgeprops={'linewidth':6}) 
plt.title("split among classes")
plt.show()
```

```python
#boxplot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
plt.show()

# here circles are outliers
# 0% , 25% , 50% , 75% , 100%
# percentile distribution
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2

# s - size of points, can be a single number or an array
# c - colour of the points
# alpha - transparency value of colours
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

```python
# historams
# histograms are useful to visualize how the data is distributed in some buckets of particular range
from random import sample
data = sample(range(1, 1000), 100) 
#here, we are sampling 100 values between 1 and 1000
#let's say we want to see how many values are sampled between 0-100 or 100-200 (buckets) and so on
#we can use histogram for this
plt.hist(data) 
#by default, it will plot the data distribution in 10 buckets and on y axis, we have the number of values in a particular bucket
```

```python
#from above, we can clearly say that there are approx 9 values sampled in the bucket (0-100) and so on
```
