Student Marks Prediction Model

```python
x = np.array(df[["number_courses","time_study"]])
y = np.array(df["Marks"])
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state = 42)
```

```python
model = LinearRegression()
model.fit(x,y)
model.score(x_test,y_test)
```

```python

# features = time_study, number_courses
features = np.array([[5.034,4]])
model.predict(features)
```

In this way we can predict the marks of a student with machine learning using python.

```python
# Function for predicting student marks
def StudentMarksPrediction(time_study, number_courses):
    """
    Predict the marks of a student.

    Args:
        time_study (float): Amount of time student spends studying.
        number_courses (int): Number of courses selected.

    Returns:
        float: Predicted marks.
    """
    features = np.array([[number_courses, time_study]])
    predicted_marks = model.predict(features)
    return predicted_marks

#Example usage
predicted_marks = StudentMarksPrediction(6.78, 2)
print(f"Predicted Marks: {predicted_marks}")
```

Now, let's test the performance of the machine learning model by giving inputs

```python
mb.deploy(StudentMarksPrediction)
```

```python
#Define input data

time_study = 6.78
number_courses = 2

#Get prediction from deployed model
prediction = modelbit.get_inference(
  region="ap-south-1",
  workspace="pondupondu",
  deployment="StudentMarksPrediction",
  data=[time_study, number_courses]
)

#Display the prediction
print(f"The predicted score is : {prediction}")
```
