## Seaborn

* distplot
* jointplot
* pairplot

```python
import seaborn as sns                 # importing seaborn library 

#getting data inbuilt

df=sns.load_dataset('tips')          # Dataset
df.head()
```

### correlation with Heatmap

```python
# compare the each feature to each other and find the percentage

df.corr()
```

```python
sns.heatmap(df.corr())          #    show the matrix into colored form with 2d scale
```

## Bivariate analysis
 * jointplot

```python
# joint plot draw for 2 features only change the kind
#to se diff changes Ex. reg,hex,resid,scatter

sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')                # Regression plot 
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hist')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

## plots for More than 2 features

* pairplot

```python
sns.pairplot(df)
```

```python
# also can classify based on data
#Example

sns.pairplot(df,hue='sex')                                 # Data is dived by sex category  with colored combo
                                                           # hue can be changed for diff features 
```

### categorical data set Analysis
* Countplot
* barplot
* boxplot

```python
# countplot 

sns.countplot(x='day',data=df)                                   # count the no of category 
```

```python
#barplots

sns.barplot(x='sex',y='tip',data=df)                    # one feature should be numeric 
```

### Box Plot
 * *boxplot show the skrewness of data set or normal distribution  ,outliers

```python
sns.boxplot(x='sex',y='tip',data=df)                #one numeric feature required
```

```python
sns.boxplot(x='day',y='tip',data=df)
```

## Matplotlib

.

```python
#normal plots 
from matplotlib import pyplot as plt
import numpy as np

x=np.arange(0,10)
y=[10,21,45,50,55,65,68,78,98,100]

plt.plot(x,y,'bo--')
plt.xlabel('Toal Bill')
plt.ylabel('Tips')
plt.title('A small over view of plt')

```
