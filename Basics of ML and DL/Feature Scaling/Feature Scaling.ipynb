```python
import numpy as np
import pandas as pd
```

# There are mainly 3 types of scalers:
## 1. scale  (most basic scaler, but don't use it practically)
## 2. MinMaxScaler
## 3. StandardScaler  (we will use this majorly)

StandardScaler follows Standard Normal Distribution (SND). 
Therefore, it makes mean = 0 and scales the data to unit variance. 
MinMaxScaler scales all the data features in the range [0, 1] or 
else in the range [-1, 1] if there are negative values in the dataset. 
This range is also called an Interquartile range.

## scale function

```python
columnNames = ['Alcohol' , 'Malic acid' , 'Ash' , 'Alcalinity of ash' , 'Magnesium' ,'Total phenols' , 'Flavanoids','Nonflavanoid phenols' ,'Proanthocyanins','Color intensity' ,'Hue' ,'OD280/OD315 of diluted wines' ,'Proline'     ]

wine = pd.read_csv("wine_data.csv" , names = columnNames)
```

```python
wine.head()
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(np.array(wine))
x_scaled
```

```python
x_scaled.mean(axis = 0)
```

```python
x_scaled.std(axis= 0)
```

```python
wine = pd.DataFrame(x_scaled , columns = columnNames)
# we converted it back to dataframe
```

```python
wine.head()
```

### MixMaxScaler


```python
wine = pd.read_csv("wine_data.csv" , names = columnNames)
```

```python
wine.head()
```

```python
min_max_scaler_object = preprocessing.MinMaxScaler()
```

```python
wine = min_max_scaler_object.fit_transform(wine)
```

```python
wine = pd.DataFrame(wine , columns = columnNames)
```

```python
wine.head()
```

## StandardScaler

```python
wine = pd.read_csv("wine_data.csv" , names = columnNames)
```

```python
wine.head()
```

```python
scaler = preprocessing.StandardScaler()
scaler.fit(wine)
wine = scaler.transform(wine)
```

```python
wine
```

```python
wine = pd.DataFrame(wine , columns = columnNames)
wine.head()
```
