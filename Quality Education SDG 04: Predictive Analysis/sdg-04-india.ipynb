## SDG 4: Quality Education

### Problem Statement:
Despite significant progress in education in India, challenges such as inequality, a lack of resources, and inadequate teaching methods persist. Many children and adults still lack access to quality education, particularly in marginalized communities. 

These challenges are exacerbated by insufficient data and analysis to inform policy and practice. This project addresses these issues by analyzing educational data to identify critical areas for intervention and propose actionable solutions to improve the quality of education in India.


### Objective of the Project:
The primary objective is to analyze educational data to identify key factors affecting educational quality in India and propose data-driven solutions to enhance educational outcomes. The specific objectives are as follows:
- By 2030, ensure all girls and boys complete free, equitable, and quality primary and secondary education in India.
- Provide access to quality early childhood development, care, and pre-primary education for all children.
- Guarantee equal access to affordable and quality technical and vocational education for all.
- Increase the number of youth and adults with relevant skills for employment and entrepreneurship.
- Eliminate gender disparities and ensure equal access to all education levels for vulnerable groups.
- Achieve literacy and numeracy for all youth and a substantial proportion of adults.
- Equip all learners with knowledge and skills for sustainable development, human rights, and global citizenship.


### Features:
Key features of the dataset will include:
- **Demographics:** Age, Gender, and Socio-economic status of students.
- **Performance Metrics:** Test scores, graduation rates, literacy rates.
- **Resource Allocation:** Availability of teachers
- **Attendance and Enrollment:** Enrollment rates, attendance records.
- **Contextual Factors:** Economic indicators, geographical data.
- **Educational Levels:** Data on pre-primary, primary, and secondary education.
- **Technological Integration:** Access to and use of educational technology and digital learning resources.
- **Special Needs Education:** Provisions for students with disabilities and special educational needs.

### Model Training

```python
# Define target variable and features
target = 'sdg_4_index_score'
features = df1.drop(columns=['sdg_4_index_score', 'statecode', 'statename', 'country'])

# Split data into features and target
X = features
y = df1[target]
```

### Feature Importance

```python
# Import necessary libraries
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Assuming X and y are already defined
# Fit a RandomForestRegressor to identify feature importance
model = RandomForestRegressor()
model.fit(X, y)

# Get feature importances
importances = model.feature_importances_
feature_names = X.columns
feature_importance = pd.DataFrame({'feature': feature_names, 'importance': importances})
feature_importance = feature_importance.sort_values(by='importance', ascending=False)

print("Feature Importances:")
print(feature_importance)

# Filter features with importance greater than 0.05
important_features = feature_importance[feature_importance['importance'] > 0.05]['feature']
print(f"Important Features: {important_features.tolist()}")

# Subset the DataFrame to include only important features
X_important = X[important_features]

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_important, y, test_size=0.2, random_state=42)
```

```python
# Train and evaluate a model using only important features
model_important = RandomForestRegressor()
model_important.fit(X_train, y_train)

# Predict and evaluate
y_pred = model_important.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error with Important Features: {mse}')
print(f'R-squared with Important Features: {r2}')
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import RandomizedSearchCV

# Define the parameter grid for GradientBoostingRegressor
param_dist = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7]
}

# Create a RandomizedSearchCV object
random_search = RandomizedSearchCV(GradientBoostingRegressor(), param_distributions=param_dist, n_iter=10, cv=5, scoring='neg_mean_squared_error', random_state=42)

# Train the model using RandomizedSearchCV
random_search.fit(X_train, y_train)

# Best parameters
print(f'Best parameters for Gradient Boosting: {random_search.best_params_}')
```

### Gradient Boosting

```python
# Predict using the best model
best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)

# Evaluate the best model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error with Optimized Gradient Boosting: {mse}')
print(f'R-squared with Optimized Gradient Boosting: {r2}')
```

### XGBoost

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Create and train an XGBoost model
xgb_model = XGBRegressor(n_estimators=300, max_depth=3, learning_rate=0.1)
xgb_model.fit(X_train, y_train)

# Predict and evaluate
y_pred_xgb = xgb_model.predict(X_test)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)

print(f'Mean Squared Error for XGBoost: {mse_xgb}')
print(f'R-squared for XGBoost: {r2_xgb}')
```

### K-Means Clustering

```python
import pandas as pd

# Assuming df contains 'statename' and 'sdg_4_index_score' columns
# Extract relevant columns
data = df[['statename', 'sdg_4_index_score']]

# Check the data
print(data.head())
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Reshape the data for clustering (SDG score as a feature)
X = data[['sdg_4_index_score']].values

# Fit K-Means
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster_KMeans'] = kmeans.fit_predict(X)

# Check cluster assignments
print(data.head())
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Create a color map
colors = sns.color_palette("husl", n_colors=len(data['Cluster_KMeans'].unique()))

# Create a plot
plt.figure(figsize=(12, 8))

# Scatter plot with color map
scatter = plt.scatter(data['sdg_4_index_score'], np.zeros_like(data['sdg_4_index_score']),
                      c=data['Cluster_KMeans'], cmap=plt.cm.get_cmap('viridis', len(data['Cluster_KMeans'].unique())),
                      marker='o', edgecolor='k')

# Add labels
for i, state in enumerate(data['statename']):
    plt.annotate(state, (data['sdg_4_index_score'].iloc[i], 0), fontsize=8, alpha=0.7)

# Add color bar
plt.colorbar(scatter, label='Cluster')

# Set titles and labels
plt.title('K-Means Clustering of States Based on SDG 4 Index Score')
plt.xlabel('SDG 4 Index Score')
plt.ylabel('Cluster (Fixed at 0 for Visualization)')

plt.grid(True)
plt.show()
```
