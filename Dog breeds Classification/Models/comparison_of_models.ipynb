# Visual Representation of accuracy scores of different deep learning models that are being applied for dog breed classification

### importing libraries


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from wordcloud import WordCloud
import plotly.graph_objects as go
```

##  LINE PLOT

```python
# Accuracy scores
accuracy_scores = [0.9090909090909091,0.9818181818181818,0.9636363636363636]
model_names = ['VGG19','ResNet50V2', 'MobileNet']

# Set the figure size
plt.figure(figsize=(8, 6))

# Create the line plot
plt.plot(model_names, accuracy_scores, marker='o', linestyle='-', color='b', linewidth=2)

# Add data labels with adjusted vertical alignment and offset
offset = 0.0001 
for i in range(len(model_names)):
    plt.text(model_names[i], accuracy_scores[i]+offset, str(accuracy_scores[i]), ha='center')

# Add labels and title
plt.xlabel('Model')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores of Different Models')

# Show the plot
plt.show()
```

## DISTRIBUTION PLOT

```python
plt.figure(figsize=(8, 6))
plt.hist(accuracy_scores, bins=10, color='b', alpha=0.7)
plt.xlabel('Accuracy Score')
plt.ylabel('Frequency')
plt.title('Distribution of Accuracy Scores')
plt.show()
```

## HEATMAP 

```python
data = np.array(accuracy_scores).reshape(1, -1)
plt.figure(figsize=(8, 6))
sns.heatmap(data, cmap='Blues', annot=True, xticklabels=model_names, yticklabels=False, cbar=False)
plt.xlabel('Model')
plt.ylabel('')
plt.title('Accuracy Scores of Different Models')
plt.show()
```

## WORLDCLOUD

```python
# Convert scores to frequencies
frequencies = {model: score for model, score in zip(model_names, accuracy_scores)}

# Generate word cloud
wordcloud = WordCloud(width=1200, height=700, background_color='white').generate_from_frequencies(frequencies)

# Plot word cloud
plt.figure(figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Accuracy Scores of Different Models')

# Show the plot
plt.show()
```

## BUBBLE CHART

```python
# Create the bubble chart
fig = go.Figure(data=go.Scatter(
    x=model_names,
    y=accuracy_scores,
    mode='markers',
    marker=dict(
        size=[score * 1000 for score in accuracy_scores],
        color=accuracy_scores,
        colorscale='Blues',
        showscale=True
    )
))

# Set title and axis labels
fig.update_layout(title='Accuracy Scores of Different Models')
fig.update_xaxes(title_text='Model')
fig.update_yaxes(title_text='Accuracy Score')

# Show the chart
fig.show()
```
